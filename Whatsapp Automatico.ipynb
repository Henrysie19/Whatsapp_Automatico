```python
# Importación de módulos
import time 
import webbrowser as web
import pyautogui as pg
import requests
import pandas as pd
from pyautogui import click, hotkey, locateOnScreen, moveTo, press, size, typewrite
from platform import system
import win32clipboard
from io import BytesIO
from PIL import Image
import cv2
import os
import string
```

```python
# Lectura de datos desde un archivo CSV
data = pd.read_csv("Pruebamensajespython.csv", encoding ='latin1')
data_dict = data.to_dict('list')
nombres = data_dict['nombre']
celulares = data_dict['celular']
mensajes = data_dict['mensaje']
numeroindicativo = '+57'
combo = zip(nombres,celulares,mensajes)

# Función para copiar una imagen al portapapeles   
def copy_image(file_name= str) -> None:
        """Copy the Image to Clipboard based on the Platform"""
        
        # ... (código para copiar la imagen al portapapel
        image = Image.open(file_name)
        output = BytesIO()
        image.convert("RGB").save(output, "BMP")
        data = output.getvalue()[14:]
        output.close()
        win32clipboard.OpenClipboard()
        win32clipboard.EmptyClipboard()
        win32clipboard.SetClipboardData(win32clipboard.CF_DIB, data)
        win32clipboard.CloseClipboard()

# Función para enviar una imagen a través de WhatsApp Web        
def send_image(file_name= str) -> None:
        """Sends the Image to a Contact or a Group based on the Receiver"""
            # ... (código para enviar la imagen a través de WhatsApp Web)

        time.sleep(25.2/4)
        copy_image(file_name=file_name)
        
        pg.hotkey("ctrl", "v")
        time.sleep(0.99/100)
        pg.press("enter")

# Bucle que recorre la lista de contactos y envía mensajes e imágenes
for nombre, celular, mensaje in combo:
    # Abre la URL de WhatsApp Web con el número de teléfono y mensaje predefinidos
    web.open("https://web.whatsapp.com/send?phone=" + numeroindicativo + str(celular) + "&text=Hola" + " " + str(nombre) + " " + str(mensaje))
    web.open("https://web.whatsapp.com/send?phone="+numeroindicativo+str(celular)+ "&text=Hola"+" "+str(nombre)+" "+str(mensaje))
    
    # Envía cada una de las imágenes al contacto o grupo
    send_image(file_name="imagen1.jpg")
    send_image(file_name="imagen2.jpg")
    send_image(file_name="imagen3.jpg")
    send_image(file_name="imagen4.jpg")
    send_image(file_name="imagen5.jpg")
    send_image(file_name="imagen6.jpg")

     # Espera y realiza algunas acciones de control
    time.sleep(2)
    pg.press('enter')
    time.sleep(5)
    pg.hotkey('ctrl','w')
```
